```python
import pandas as pd
import quandl, math
import numpy as np
from sklearn import preprocessing , svm , model_selection
from sklearn.linear_model import LinearRegression
```

```python
df1=quandl.get('WIKI/GOOGL')
```

```python
df1=df1[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume',]]
df1.head()
```

```python
df1['hc_chng']=(df1['Adj. High']-df1['Adj. Close'])/df1['Adj. Close']*100.0
df1['hl_chng']=(df1['Adj. High']-df1['Adj. Low'])/df1['Adj. Low']*100.0
df1['oc_chng']=(df1['Adj. Close']-df1['Adj. Open'])/df1['Adj. Open']*100.0
df1=df1[['Adj. Close','hc_chng','hl_chng','oc_chng','Adj. Volume']]
df1.head()
```

```python
forecast='Adj. Close'
df1.fillna(-99999,inplace=True)
forecast_len=int(math.ceil(0.01*len(df1)))
df1['label']=df1[forecast].shift(-forecast_len)
print('DAY SKIP=',forecast_len,' TOTAL LEN=',len(df1))
df1.dropna(inplace=True)
df1.head()
```

```python
X=np.array(df1.drop(['label'],1))
y=np.array(df1['label'])
X=preprocessing.scale(X)
print(len(X))
print(X)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
clf=LinearRegression()
# clf=svm.SVR(kernel='poly')
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(accuracy)
# clf.score(X_train, y_train)
```
